```python
import pandas as pd
import json
import numpy as np
```

```python
DATA_FOLDER = 'data/'
DICTIONNARY_FOLDER = DATA_FOLDER + 'dictionnaries/'
DICT_DEGREES = DICTIONNARY_FOLDER + 'academic_degrees_categories.csv'
```

```python
data = pd.read_csv(DICT_DEGREES, sep = ';', error_bad_lines=False)
data.head()
```

```python
columns = data.columns.tolist()
```

```python
data_dictionnary = {}
```

```python
for col in columns:
    keys = data[col].dropna().tolist()
    for k in keys:
        data_dictionnary[k] = col
```

```python
data_dictionnary
```

```python
with open('degrees_categories.json', 'w') as json_file:
    json.dump(data_dictionnary, json_file)
```
