```python
import pandas as pd
import bz2
import json
```

```python
DATA_FOLDER = './'
SPEAKER_ATTRIBUTES_DATA = DATA_FOLDER+"speaker_attributes.parquet"

KNOWN_SPEAKER = DATA_FOLDER+"100000_knownspeakers_2019"
KNOWN_SPEAKER_DATA = KNOWN_SPEAKER+".json.bz2"
RETURN_FILE_DATA = KNOWN_SPEAKER+"_with_attributes.json.bz2"
```

```python
#Read the parquet file and the dataframe (sampled as wished beforehand) to add the speakers attributes to
parquet = pd.read_parquet(SPEAKER_ATTRIBUTES_DATA)
df = pd.read_json(KNOWN_SPEAKER_DATA)
```

```python
#Only the first speaker of each citations (the one with higher probability) is looked at
df['qids'] = df['qids'].apply(lambda x: x[0])
```

```python
#Create a new dataframe with only the informations needed from the parquet file
parquet_new = parquet.get(['gender','date_of_birth','occupation','id'])
parquet_new.rename(columns={'id':'qids'}, inplace=True)
```

```python
#Create the new dataframe with the additionnal informations about the speakers (gender, age, occupation)
df_new = df.merge(parquet_new)
df_new
```

```python
#Creation of a json file from df_new
df_new.to_json(RETURN_FILE_DATA)
```
